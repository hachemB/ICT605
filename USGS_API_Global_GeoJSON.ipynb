# Importing Data

```python
# Define the query parameters
query = dict(starttime="2013-09-22", endtime="2023-09-22", minmagnitude="5.0", format="geojson", limit="20000")
query_string = '&'.join('{0}={1}'.format(k, v) for k, v in query.items())
query_url = "http://earthquake.usgs.gov/fdsnws/event/1/query?" + query_string

# Get the data
response = requests.get(query_url)
data = response.json()

# Convert the GeoJSON data to a GeoDataFrame
gdf = gpd.GeoDataFrame.from_features(data['features'])
```

```python
# Categorizing earthquakes based on magnitude
categories = ["Moderate", "Strong", "Major", "Great"]
bins = [5.0, 6.0, 7.0, 8.0, np.inf]
gdf['magnitude_category'] = pd.cut(gdf['mag'], bins=bins, labels=categories, right=False)

# Determining the frequency for each category
frequency = gdf['magnitude_category'].value_counts().sort_index()

# Calculating the percentage for each category
total_earthquakes = len(gdf)
percentage = (frequency / total_earthquakes) * 100

# Displaying the frequency and percentage
for cat, freq, perc in zip(categories, frequency, percentage):
    print(f"{cat}: {freq} earthquakes ({perc:.2f}%)")

# Optional: Plotting the frequencies with percentage annotations
plt.figure(figsize=(10, 6))
ax = frequency.plot(kind='bar', edgecolor='black')
plt.title('Frequency of Earthquakes in Different Richter Scale Categories')
plt.xlabel('Magnitude Category')
plt.ylabel('Number of Earthquakes')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Annotating the bars with percentages
for p in ax.patches:
    ax.annotate(f"{p.get_height()} ({p.get_height() / total_earthquakes * 100:.2f}%)", 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 10), 
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

# Map distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import shapely.geometry

# Adjust the longitude values in the GeoDataFrame
gdf['geometry_map'] = gdf['geometry'].apply(lambda geom: geom if geom.x >= 0 else shapely.geometry.Point(geom.x + 360, geom.y))

# Create a new figure with a specific size
plt.figure(figsize=(15, 10))

# Create a basemap with 'mill' projection (cylindrical)
m = Basemap(projection='mill',llcrnrlat=-60,urcrnrlat=90,\
            llcrnrlon=0,urcrnrlon=360,resolution='c')

# Draw coastlines and countries
m.drawcoastlines()
m.drawcountries()

# Extract latitude and longitude from the GeoDataFrame
latitude = gdf['geometry_map'].y.values
longitude = gdf['geometry_map'].x.values

# Convert latitude and longitude to x and y coordinates
x, y = m(longitude, latitude)

# Plot the heatmap using hexbin
m.hexbin(x, y, gridsize=500, bins='log', cmap='inferno')

# Add a colorbar on the right side
cb = plt.colorbar(label='Number of Earthquakes')
cb.set_label('Log(Number of Earthquakes)')

# Add a title
plt.title("Earthquake Activities (Magnitude > 5.0) from 2013-09-22 to 2023-09-22")

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns

# Create a new column for the month
gdf['month'] = gdf.index.month

# Bin the magnitudes
bins = [5, 6, 7, 8, 9, 10]
labels = ['5-6', '6-7', '7-8', '8-9', '9-10']
gdf['magnitude_bin'] = pd.cut(gdf['mag'], bins=bins, labels=labels, right=False)

# Create a pivot table for the heatmap
heatmap_data = gdf.pivot_table(index='magnitude_bin', columns='month', aggfunc='size', fill_value=0)

# Month names for labeling the x-axis
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Plotting the heatmap with adjusted annotations
plt.figure(figsize=(12, 7))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', linewidths=.5, xticklabels=months, fmt='d', cbar_kws={'label': 'Earthquake Count'})
plt.title('Earthquake Frequency by Month and Magnitude')
plt.xlabel('Month')
plt.ylabel('Magnitude')
plt.tight_layout()
plt.show()

```

```python
import numpy as np

# Calculate the earthquake counts by magnitude bin for each month
monthly_counts_by_bin = gdf.groupby(['month', 'magnitude_bin']).size().unstack(fill_value=0)

# Calculate the monthly average magnitude
monthly_avg_magnitude = gdf.groupby('month')['mag'].mean()

# Set up the plot
fig, ax1 = plt.subplots(figsize=(15, 7))

# Plot the bars (monthly counts by magnitude bin)
monthly_counts_by_bin.plot(kind='bar', stacked=True, ax=ax1, colormap='viridis')
ax1.set_ylabel('Earthquake Count')
ax1.set_xlabel('Month')
ax1.set_title('Monthly Earthquake Frequency and Average Magnitude')
ax1.set_xticklabels(months, rotation=45)

# Set up the secondary Y-axis for the line plot
ax2 = ax1.twinx()
ax2.plot(monthly_avg_magnitude.index, monthly_avg_magnitude.values, color='red', marker='o', label='Average Magnitude')
ax2.set_ylabel('Average Magnitude', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
